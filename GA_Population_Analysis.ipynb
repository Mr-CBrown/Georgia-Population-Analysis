```python
#import packages 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn 

ga = pd.read_csv('georgiadata.csv') #import csv file with pandas
```

```python
print(ga.head()) # view the file to confirm data properly transfered
```

```python
#Pull out the data we want for our analysis
latitude, longitude = ga['latitude'], ga['longitiude'] #create variable for the lat and long 
population, area = ga['pop2022'],ga['aland_sqmi'] # create variables for the population and area of land
```

```python
seaborn.set()
plt.scatter(longitude, latitude, label=None, c=np.log10(population),
            cmap='viridis', s=area, linewidth=0, alpha=0.5)
plt.axis('equal')
plt.xlabel('Longitude')
plt.ylabel('Longitude')
plt.colorbar(label='log$_{10}$(population)')
plt.clim(3, 7)
# Create a legend, we will plot empty lists with the desired size and label
for area in [50, 100, 250]:
    plt.scatter([], [], c='k', alpha=0.3, s=area, label=str(area) + 'mi$^2$')
plt.legend(scatterpoints=1, frameon=False, labelspacing=1, title='City Areas')
plt.title("Area and Population of Georgia Cities")
plt.rcParams['figure.figsize'] = [10/5, 8/5]
plt.show()
```
